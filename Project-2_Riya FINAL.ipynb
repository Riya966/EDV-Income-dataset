## 5. Check Null values, outliers and duplicate data and perform data cleaning


```python
## Cheking the null values and duplicate data from the dataset.
print("Null values per columns are:\n",data1.isnull().sum())
print()
print("Duplicate data count:",data1.duplicated().sum())
```

```python
## Replacing the ? with the nan and checking out the how many null values are there.
data1.replace('?',np.nan,inplace=True)
print("Null values per columns are:\n",data1.isnull().sum())
```

```python
## cleaning the dataset thorugh dropping the null and duplicate values from the dataset.
data1.dropna(inplace=True)
data1.drop_duplicates(inplace=True)
```

```python
## checking if all the null values and duplicate values are dropped or not. 
print("Null values per columns are:\n",data1.isnull().sum())
print()
print("Duplicate data count:",data1.duplicated().sum())
print(data1.shape)
```

```python
## checking for the outliers in the age column.
plt.figure(figsize=(8,5))
sns.boxplot(data=data1['age'])
plt.title("box plot to detect outliers")
plt.ylabel("values")
plt.show()
```

```python
## finding min and max values from the age column.
print('MIN AGE:',data1['age'].min(),'MAX AGE:',data1['age'].max())
```

```python
## finding lower and upper bound value using IQR method.
Q1 = data1['age'].quantile(0.25)
Q3 = data1['age'].quantile(0.75)
IQR = Q3 - Q1
print("IQR is :",IQR)

lower_bound = Q1 - 1.5 * IQR  ## if we are not finding the lower bound that is also fine as age can't be negative
upper_bound = Q3 + 1.5 * IQR
print(f"Lower bound is: {lower_bound} and Upper Bound is {upper_bound} ")
```

```python
## creating dataframe of all the outlier values.
Outliers_age = pd.DataFrame(data1[(data1['age'] < lower_bound) | (data1['age'] > upper_bound)])
print("outlier detected:")
Outliers_age
```

```python
## cheking for outliers in the hours-per-week column using BOX plot.
plt.figure(figsize=(8,5))
sns.boxplot(data=data1['hours-per-week'])
plt.title("box plot to detect outliers in the Hours-per-week column")
plt.ylabel("values")
plt.show()
```

```python
## finding maximum and minimum value.
print('Min Hours per week:',data1['hours-per-week'].min(),'Max Hours per week:',data1['hours-per-week'].max())
```

```python
## finding lower and upper bound using IQR method.
Q1 = data1['hours-per-week'].quantile(0.25)
Q3 = data1['hours-per-week'].quantile(0.75)
IQR = Q3 - Q1
print("IQR is :",IQR)

lower_bound = Q1 - 1.5 * IQR  ## if we are not finding the lower bound that is also fine as age can't be negative
upper_bound = Q3 + 1.5 * IQR
print(f"Lower bound is: {lower_bound} and Upper Bound is {upper_bound} ")
```

```python
## creating dataframe of all the outliers.
Outliers_hours_per_week = pd.DataFrame(data1[(data1['hours-per-week'] < lower_bound) | (data1['hours-per-week'] > upper_bound)])
print("outlier detected:")
Outliers_hours_per_week
```

```python
## cleaning the data by removing outliers and have filter out the records outside the bounds.
data_cleaned = data1[(data1['hours-per-week'] >= 32.5) & (data1['hours-per-week'] <= 52.5)]
data_cleaned
```

## 12. Who Has Better Chance To Get Salary >50K Male or Female ?

```python
## filter the data where income >50k
high_income=data1[data1['income']=='>50K']

## counts the gender
gender_count = data1['gender'].value_counts()
high_income_gender_count= high_income['gender'].value_counts()

## printing both the values
print("The Total count of male and female in the dataframe:\n",gender_count)
print("\nIncome >50k count in male and female:\n",high_income_gender_count)
```

```python
## Calculate the proportion of >50K earners within each gender

proportion = (high_income_gender_count/gender_count)*100
print("Proportion of >50K earners by gender:\n",proportion)

## finding for better chance gender

better_chance_gender = proportion.idxmax()
print("\nGender with better chance to earn >50k is:",better_chance_gender)
print()
## plot the data using pie chart.
proportion.plot(kind='pie',autopct='%0.1f%%')
plt.legend()
```
